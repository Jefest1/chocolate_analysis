## importing necessary modules and packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
cocoa_df = pd.read_excel("flavors_of_cacao.xlsx")
cocoa_df.head(20)
```

```python
# Getting all the columns
cocoa_df.columns
```

```python
#Renaming the columns to suit the dataframe
cocoa_df = cocoa_df.rename(columns = {'Company??\n(Maker-if known)': 'Company Name',
                           'Specific Bean Origin\nor Bar Name': 'Specified Bean Origin',
                           'Review\nDate': 'Review Year',
                           'Cocoa\nPercent': 'Cocoa Percent',
                           'Company\nLocation': 'Company Location',
                           'Bean\nType': 'Bean Type',
                           'Broad Bean\nOrigin': 'Country'})
cocoa_df
```

```python
# Making some descriptive statistics on the data
cocoa_df.describe()
```

```python
# checking for null values in the data
cocoa_df.isna().sum()
```

```python
#checking for duplicated values and removing them
cocoa_df.loc[cocoa_df.duplicated()]
```

### Exploratory Analysis

```python
cocoa_df.columns
```

```python
# Top 50 cocoa Producing countries
countries = cocoa_df["Country"].value_counts().head(10)
countries
```

#### Visualizing

```python
# Create a bar plot
plt.bar(countries.index, countries.values)

# Set the title and labels for the plot
plt.title("Top 10 Cocoa Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Cocoa Production")
plt.xticks(rotation=45)
plt.show()
```

```python
# create a plot for corelation
cor = cocoa_df[['Review Year', 'Cocoa Percent', 'Rating']].corr()
cor
```

```python
sns.heatmap(cor, annot=True, cmap='coolwarm')

# Set the title for the plot
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
