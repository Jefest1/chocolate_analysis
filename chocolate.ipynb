## importing necessary modules and packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
cocoa_df = pd.read_excel("flavors_of_cacao.xlsx")
cocoa_df.head(20)
```

```python
# Getting all the columns
cocoa_df.columns
```

```python
#Renaming the columns to suit the dataframe
cocoa_df = cocoa_df.rename(columns = {'Company??\n(Maker-if known)': 'Company Name',
                           'Specific Bean Origin\nor Bar Name': 'Specified Bean Origin',
                           'Review\nDate': 'Review Year',
                           'Cocoa\nPercent': 'Cocoa Percent',
                           'Company\nLocation': 'Company Location',
                           'Bean\nType': 'Bean Type',
                           'Broad Bean\nOrigin': 'Country'})
cocoa_df
```

```python
# Making some descriptive statistics on the data
cocoa_df.describe()
```

```python
# checking for null values in the data
cocoa_df.isna().sum()
```

```python
#checking for duplicated values and removing them
cocoa_df.loc[cocoa_df.duplicated()]
```

### Exploratory Analysis

```python
cocoa_df.columns
```

```python
# Top 50 cocoa Producing countries
countries = cocoa_df["Country"].value_counts().head(10)
countries
```

#### Visualizing

```python
# Create a bar plot
plt.bar(countries.index, countries.values)

# Set the title and labels for the plot
plt.title("Top 10 Cocoa Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Cocoa Production")
plt.xticks(rotation=45)
plt.show()
```

```python
# create a plot for corelation
cor = cocoa_df[['Review Year', 'Cocoa Percent', 'Rating']].corr()
cor
```

```python
sns.heatmap(cor, annot=True, cmap='coolwarm')

# Set the title for the plot
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```

```python
# top 10 chocolate producing companies
#add a distribution chart
Companies = cocoa_df['Company Name'].value_counts().head(10)
Companies
```

```python
sns.distplot(Companies.values)

# Set the title and labels for the plot
plt.title("Distribution of Top 10 Chocolate Producing Companies")
plt.xlabel("Number of Chocolate Productions")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
# Plot scatterplot of cocoa percentage and ratings
sns.scatterplot(data=cocoa_df, x="Cocoa Percent", y="Rating")
plt.title("Relationship between Cocoa Percentage and Chocolate Ratings")
plt.xlabel("Cocoa Percentage")
plt.ylabel("Rating")
plt.show()
```

```python
# Get the top 10 chocolate companies by number of reviews
top_companies = cocoa_df["Company Name"].value_counts().head(10).index.tolist()

# Subset the dataframe to only include top companies
top_chocolate = cocoa_df[cocoa_df["Company Name"].isin(top_companies)]

# Group the top chocolate by company and compute the mean rating
mean_rating = top_chocolate.groupby("Company Name")["Rating"].mean().sort_values(ascending=False)

# Plot bar chart of mean ratings for top chocolate companies
sns.barplot(x=mean_rating.values, y=mean_rating.index)
plt.title("Mean Ratings for Top Chocolate Companies")
plt.xlabel("Mean Rating")
plt.ylabel("Company Name")
plt.show()
```
